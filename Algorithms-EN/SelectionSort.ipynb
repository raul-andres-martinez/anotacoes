# Selection Sort

Selection sort is a sorting algorithm that works by repeatedly finding the minimum (or maximum) element in the unsorted portion of the list and swapping it with the first element of that unsorted portion. This process continues until the entire list is sorted.

## Time and Space Complexity

**Time Complexity: O(n^2).** In the worst case scenario, each element is compared with every other element, leading to n*(n-1)/2 comparisons which is approximately O(n^2).

**Space Complexity: O(1).** Selection sort is an in-place algorithm, meaning it sorts the data within the original array without requiring significant additional space. It only uses a few temporary variables for comparisons and swapping.

```python
import matplotlib.pyplot as plt
import numpy as np

n = np.arange(1, 101)

time_complexity = n**2

plt.figure(figsize=(8, 6))
plt.plot(n, time_complexity, label="O(n^2)")
plt.xlabel("Input size (n)")
plt.ylabel("Comparisons")
plt.title("Time complexity - O(n^2)")
plt.legend()
plt.grid(True)
plt.show()
```

## Implementation
The implementation uses nested loops to achieve the sorting. The outer loop iterates through the array, and for each element, the inner loop finds the index of the minimum element in the remaining unsorted part of the array. Finally, the elements at those indices are swapped.

```python
def selectionSort(arr):
    for i in range(0, len(arr)-1):
        # Set lowest index from the unsorted part
        low = i
        for k in range(i+1, len(arr)):
            # Set new lowest element
            if arr[k] < arr[low]:
                low = k
        # Swap the new lowest element with the first element from the unsorted part
        arr[i], arr[low] = arr[low], arr[i]
    print(arr)

# Test example        
print(selectionSort([4, 2, 10, 1, 0, 9, 5]))
```
